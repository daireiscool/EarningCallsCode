```python
import pandas as pd
import numpy as np
import plotly.express as px
import json
```

```python
df = pd.read_csv("MetricsOfEarningCallsV2.csv", index_col=0)
```

```python
df["Date"] = df["year"].astype(str) + "-" + ((df["quarter"]*3) - 1).astype(str).apply(lambda x: x.zfill(2)) + "-01"
df["Date"] = pd.to_datetime(df["Date"])
df["Company"] = df["symbol"].map({
    "AAPL":"Apple",
    "GOOG":"Google",
    "AON":"Aon",
    "C":"Citigroup Inc",
    "CVX":"Chevron Corp.",
    "GS":"Goldman Sachs",
    "JNJ":"Johnson and Johnson",
    "JPM":"J.P. Morgan",
    "META":"META",
    "MSFT":"Microsoft",
    "UNH":"United Healthcare",
    "XOM":"Exxon Mobil Corp",})
df["Type"] = df["symbol"].map({
    "AAPL":"Tech",
    "GOOG": "Tech",
    "AON":"Financial",
    "C":"Financial",
    "CVX":"Other",
    "GS":"Financial",
    "JNJ":"Other",
    "JPM":"Financial",
    "META":"Tech",
    "MSFT":"Tech",
    "UNH":"Other",
    "XOM":"Other",})
```

```python
dfMean = df[df["index"] == "max"]
dfMean = dfMean[dfMean["year"] > 2013]
```

```python
dfMean[['Company', 'Type']].drop_duplicates()["Type"].value_counts()
```

```python
fig = px.line(dfMean, x = "Date", y = "gpt_zero_Label", line_group = "symbol", color = "Type", title = "GPTZero")
fig.show()

fig = px.line(dfMean, x = "Date", y = "roberta_Label", line_group = "symbol", color = "Type", title = "RoBERTa base")
fig.show()

fig = px.line(dfMean, x = "Date", y = "roberta_large_Label", line_group = "symbol", color = "Type", title = "RoBERTa large")
fig.show()

fig = px.line(dfMean, x = "Date", y = "gltr_Label", line_group = "symbol", color = "Type", title = "GLTR")
fig.show()
```

```python
dfMean_ = dfMean.groupby(["Type", "Date"]).mean().reset_index()
```

```python
fig = px.line(dfMean_, x = "Date", y = "gpt_zero_Label", color = "Type", title = "GPTZero")
fig.show()

fig = px.line(dfMean_, x = "Date", y = "roberta_Label", color = "Type", title = "RoBERTa base")
fig.show()

fig = px.line(dfMean_, x = "Date", y = "roberta_large_Label", color = "Type", title = "RoBERTa large")
fig.show()

fig = px.line(dfMean_, x = "Date", y = "gltr_Label", color = "Type", title = "GLTR")
fig.show()
```

```python
fig = px.box(dfMean, x = "Date", y = "gpt_zero_Label", color = "Type", title = "GPTZero")
fig.show()

fig = px.box(dfMean, x = "Date", y = "roberta_Label", color = "Type", title = "RoBERTa base")
fig.show()

fig = px.box(dfMean, x = "Date", y = "roberta_large_Label", color = "Type", title = "RoBERTa large")
fig.show()

fig = px.box(dfMean, x = "Date", y = "gltr_Label", color = "Type", title = "GLTR")
fig.show()
```
